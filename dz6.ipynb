В магазин привезли устройства с 3-х разных предприятий.
Соотношение устройств следующее: 20% - продукция первого предприятия, 30% - продукция второго предприятия, 50% - продукция третьего предприятия; далее, 10% продукции первого предприятия высшего сорта, на втором предприятии - 5% и на третьем - 20% продукции высшего сорта.
Найти вероятность того, что случайно купленная новая продукция окажется высшего сорта.
Используя формулу полной вероятности, можно вычислить вероятность того, что случайно купленная новая продукция окажется высшего сорта:

P(H) = P(H|A) * P(A) + P(H|B) * P(B) + P(H|C) * P(C),

где P(H) - искомая вероятность того, что продукция окажется высшего сорта, P(A), P(B) и P(C) - вероятности купить продукцию соответствующего предприятия, P(H|A), P(H|B) и P(H|C) - условные вероятности покупки продукции высшего сорта соответствующего предприятия.

P(A) = 0.2, P(B) = 0.3, P(C) = 0.5,
P(H|A) = 0.1, P(H|B) = 0.05, P(H|C) = 0.2.

Подставим значения:

P(H) = 0.1 * 0.2 + 0.05 * 0.3 + 0.2 * 0.5 = 0.105.

Таким образом, вероятность того, что случайно купленная новая продукция окажется высшего сорта, составляет 10.5%.


Пусть брошены 3 уравновешенные монеты.

Рассмотрим 3 события:

A1 - монеты 1 и 2 упали одной и той же стороной;

A2 - монеты 2 и 3 упали одной и той же стороной;

A3 - монеты 1 и 3 упали одной и той же стороной.

Покажите, почему эти 3 события (A1, A2, A3) являются попарно независимыми, но не являются независимыми в совокупности.
Для того, чтобы события A1, A2 и A3 были попарно независимыми, нужно чтобы наличие или отсутствие одного из них не влияло на вероятность другого события. Например, для событий A1 и A2, если монета 2 выпала орлом, то события A1 и A2 не могут произойти одновременно, поэтому они зависимы. Однако, если монета 2 выпала решкой, то события A1 и A2 могут произойти одновременно, поэтому они независимы. Аналогично, события A1 и A3, A2 и A3 также являются попарно независимыми.

Однако, события A1, A2 и A3 не являются независимыми в совокупности, так как наличие или отсутствие одного из них влияет на вероятность других событий. Например, если мы знаем, что событие A1 произошло, то вероятность событий A2 и A3 возрастает, так как уже две монеты упали одной и той же стороной.
